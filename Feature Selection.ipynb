### Removing irrelevant features for classification

```python
from sklearn.feature_selection import SelectKBest , chi2 , f_classif
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
features = iris.data.astype(int)
target = iris.target
```

```python
chi2_selector = SelectKBest(chi2,k=2)
```

### K = Hyperparameter and chi2 is selecting columns with most impact on model

```python
chi2_selector.fit_transform(features,target)
```

```python
from sklearn.feature_selection import SelectFromModel
```

```python
from sklearn.svm import LinearSVC
```

```python
svc = LinearSVC().fit(iris.data,iris.target)
```

```python
 model = SelectFromModel(svc,prefit =True)
```

```python
model.transform(iris.data)
```

### PCA gives ability to derive features from multiple features

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
x = pca.fit_transform(iris.data)
```

```python
svc.fit(x,iris.target)
```

```python
A = svc.predict(x)
```

```python
from sklearn.metrics import accuracy_score 
```

```python
print(accuracy_score(A,iris.target))
```

### Variance Threshold

### Will remove columns with variance less than .5 and columns will not make much impact as it doesnt vary a lot with other columns

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
vr_thresh = VarianceThreshold(threshold=.5)
```

```python
import numpy as np
```

```python
vr_thresh.fit_transform(iris.data)
```
