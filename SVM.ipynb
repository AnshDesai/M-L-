```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
```

```python
wine = load_iris()
```

```python
data = pd.DataFrame(wine.data)
```

```python
data.head()
```

```python
y = wine.target
```

```python
X = data
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.svm import LinearSVC
```

```python
svm_class = LinearSVC(C=2,random_state=10)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
svm_class.fit(X_train,y_train)
```

```python
y_pred = svm_class.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)*100
```

```python
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(10000,noise=.1)
```

```python
y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1],c=y, s=50);
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(2)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_fit = poly.fit_transform(X)
X_train,X_test,y_train,y_test = train_test_split(X_fit,y)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
### Here svm will take kernel as argument where kernel ='linear' means simply Logistic Regression and if we take rbf(default value) then it will look out for all possible degrees.0 
```

```python
from sklearn.svm import SVC
```

```python
X_train1,X_test1,y_train1,y_test1 = train_test_split(X,y)
```

```python
svm_class = SVC(kernel = 'linear')
```

```python
svm_class.fit(X_train1,y_train1)
```

```python
svm_class.score(X_test1,y_test1)
```

```python
svm_class = SVC(kernel = 'rbf')
svm_class.fit(X_train1,y_train1)
svm_class.score(X_test1,y_test1)
```
